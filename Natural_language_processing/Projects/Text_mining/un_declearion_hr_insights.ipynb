### 1 & 2

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords

# Loading the text document and reading imto a datastructure
with open('Datasets/un_declaration_hr_text_data.txt', 'r') as file:
    documentTxt = file.read()

# Tokenizing the content in the file into words
words = documentTxt.split()

# Excluding stop words
stopWrds = set(stopwords.words('english'))
words = [word for word in words if word.lower() not in stopWrds]

# Generate word frequencies
wordFrq = {}
for word in words:
    wordFrq[word] = wordFrq.get(word, 0) + 1 #incremented by 1

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(wordFrq)

# Display the word cloud
plt.figure(figsize=(18, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

#saving the diagram as an image
wordcloud.to_file('Output/word_cloud.png')
```

### 3

```python
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

#fucntion to read the text document as a string
def readDoc(file_path):
    with open(file_path, 'r') as file:
        document = file.read()
    return document

#Function to generate bar plot
def genBar(document):
    # Tokenize the document into words
    words = word_tokenize(document)

    # Load the stop words
    stopWrds = set(stopwords.words('english'))

    # Remove stop words from the word list
    words = [word.lower() for word in words if word.lower() not in stopWrds]

    # Get the frequency count of words
    wordCnt = Counter(words)

    # Get the top 25 frequent terms
    topFreq = wordCnt.most_common(25)

    # Extract the terms and their counts
    terms, counts = zip(*topFreq)

    # Generate the bar plot
    plt.figure(figsize=(12, 6))
    plt.bar(terms, counts)
    plt.xlabel('Terms')
    plt.ylabel('No. of Frequencies')
    plt.xticks(rotation=45)
    plt.title('Non-stop words Frequent Terms (Top 25)')
    plt.tight_layout()
    plt.show()

# Read the document as a string
file = 'Datasets/un_declaration_hr_text_data.txt' #file path
doc = readDoc(file) #calling dunction and containing it in a palceholder

# Generate the bar plot by calling back function
genBar(doc)

# Saving the diagram as an image
plt.savefig('Output/most_freq_terms.png')
```
